# Identifying objects from images

The CIFAR-10 dataset contains colour images of objects, each 32x32x3 pixels (for the three colour channels). 

These have 10 categories (or classes) of object (airplane, automobile, bird, cat, deer, dog, frog, horse, ship & truck) with 5,000 images in each, making a total of 50,000 images in the training set (x_train, y_train), randomly ordered with numerical labels for each (1=airplane, 2=automobile etc.). 

The test set (x_test, y_test) contains 10,000 images ordered by their label.


```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Prepare the data

```python
# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255  # (50000, 32, 32, 3)
x_test = x_test.astype("float32") / 255 # (10000, 32, 32, 3)

# Need an extra dimension for colour channels
print("x train shape", x_train.shape)
print("x test shape", x_test.shape)


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes) # (50000, 10)
y_test = keras.utils.to_categorical(y_test, num_classes) # (10000, 10)

print("y train shape:", y_train.shape)
print("y test shape:", y_test.shape)
```

# Build the model

```python
# Model / data parameters
num_classes = 10
input_shape = (32, 32, 3)

model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        # l1 
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu", padding = 'same'),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu", padding = 'same'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(rate=0.25),
        # l2
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu", padding = 'same'),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu", padding = 'same'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(rate=0.25),
        # fc1
        layers.Flatten(),
        layers.Dense(512, activation="relu"),
        layers.Dropout(rate=0.5),
        # fc2 
        layers.Dense(10, activation="softmax"),
    ]
)


model.summary()
```

optimizers

- RMSprop: the RMSprop can solve the problem on learning rate decay  (sum of square)

- Adadelta (leanring rate is dynamic )

- dropout: dropout only affect the training set, but not affect when doing the validation
- data augmentation: we do shuffle on the training data to make it more generalized, but we didn't do the same when testing
